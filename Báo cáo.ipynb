## Linear Regression

### Giới thiệu

Hồi quy tuyến tính là một kỹ thuật phân tích dữ liệu dự đoán giá trị của dữ liệu không xác định bằng cách sử dụng một giá trị dữ liệu liên quan và đã biết khác. Nó mô hình toán học biến không xác định hoặc phụ thuộc và biến đã biết hoặc độc lập như một phương trình tuyến tính. \
Trong thuật toán Regression, chúng ta sẽ tìm 1 hàm số phù hợp với training set.

<div style="text-align:center">

Chúng ta sẽ dựa vào hàm số này để dự đoán output, và vì thuật toán này thuộc nhóm Supervised Learning, chúng ta có thể hoàn toàn nắm được mối quan hệ giữa input và output thông qua hàm số dự đoán: f(x)    



### Hàm dự đoán f(x)

Hàm f(x) là hàm dùng để dự đoán output trong thuật toán regression. Có thể xem f(x) như một công thức đặc biệt dùng để “tính” ra output. Công thức này còn được gọi là model cho thuật toán Machine Learning .   

<div >
    <p style="text-align:center">f(x) = $w_0 + w_1x_1 + w_2x_2 + ... + w_nx_n$</p>
</div>

trong đó, w1,w2,w3,w0 là các hằng số, w0 còn được gọi là bias. Mối quan hệ y≈f(x)bên trên là một mối quan hệ tuyến tính (linear). Bài toán chúng ta đang làm là một bài toán thuộc loại regression. Bài toán đi tìm các hệ số tối ưu {w1,w2,w3,w0}chính vì vậy được gọi là bài toán Linear Regression.  
Trong phương trình (1), nếu chúng ta đặt w = [w0,w1,..,wn] là vector hệ số cần phải tối ưu và x = [1,x1,x2,...,xn] là vector dữ liệu đầu vào mở rộng, Khi đó chương trình có thể viết dưới dạng:  

<div >
    <p style="text-align:center">$y \sim Xw = y'$</p>
</div>


### Sai số dự đoán

Chúng ta mong muốn rằng sự sai khác e giữa giá trị thực y và giá trị dự đoán y' là nhỏ nhất. Nói cách khác, chúng ta muốn giá trị sau đây càng nhỏ càng tốt:  
<div >
    <p style="text-align:center">$\frac{1}{2}e^2 = \frac{1}{2}(y -y')^2 = \frac{1}{2}(y-xw)^2$</p>
</div>


### Hàm mất mát

Hàm số L(w) được gọi là hàm mất mát (loss function) của bài toán Linear Regression.Chúng ta luôn mong muốn rằng sự mất mát (sai số) là nhỏ nhất, điều đó đồng nghĩa với việc tìm vector hệ số w sao cho giá trị của hàm mất mát này càng nhỏ càng tốt. Giá trị của w làm cho hàm mất mát đạt giá trị nhỏ nhất được gọi là điểm tối ưu (optimal point):  
<div >
    <p style="text-align:center">L(w) = $\frac{1}{2}$ $\sum_{i=1}^{n}$  $( y_i - x_iw)^2$</p>
</div>


### Nghiệm cho bài toán Linear Regression

Cách phổ biến nhất để tìm nghiệm cho một bài toán tối ưu là giải phương trình đạo hàm bằng 0! Tất nhiên đó là khi việc tính đạo hàm và việc giải phương trình đạo hàm bằng 0 không quá phức tạp. Thật may mắn, với các mô hình tuyến tính, hai việc này là khả thi.  
Đạo hàm theo w của hàm mất mát là:  
<div >
    <p style="text-align:center">$\frac{\partial l(w)}{\partial w}$ = $X^T(Xw - y)$</p>
</div>
Điểm tối ưu của bài toán Linear Regression có dạng:  
<div >
    <p style="text-align:center">w = $(X^TX)^\dagger$ $X^T$y</p>
</div>


### Bài toán thực tế


Chúng ta có bảng thống kê về lợi nhuận của 1 xe bán thức ăn theo dân số trong 1 thành phố như sau:

Bài toán đặt ra là: liệu có thể dự đoán mức lợi nhuận của một xe bán thức ăn dựa vào số dân số trong 1 thành phố không? (Trên thực tế, tất nhiên là không, vì lợi nhuận còn phụ thuộc vào nhiều yếu tố khác nữa). Vì vậy ta có thể giả sử là có thể dự đoán được theo Machine Learning.  


#### Hiển thị dữ liệu trên đồ thị

```python
# Thư viện dùng để phục vụ bài toán
from __future__ import division, print_function, unicode_literals
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
```

Tiếp theo, ta sẽ load đủ data vào X,y và biểu diễn các điểm trên đồ thị.  



```python
_X = np.loadtxt("univariate.txt",delimiter=",")
y = np.copy(_X[:,-1])
X = np.copy(_X[:,0])
X = X.reshape(X.size,1)
y = y.reshape(y.size,1)
plt.plot(X, y, 'ro')
plt.axis([0, 25, -6, 25])
plt.xlabel('Dân số(chục ngàn người)')
plt.ylabel('Lợi nhuận(triệu VND)')
plt.show()
```

Từ đồ thị trên ta có thể dự đoán đồ thị có dạng đường thẳng,công thức sẽ có dạng:  
(Lợi nhuận) = $w_1(dân số) + w_0$

#### Nghiệm theo công thức

Tiếp theo, ta có thể tìm ra nghiệm w1 và w0 theo công thức 5 

```python
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis = 1)

#Tính toán theo công thức
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
print('w_0 = ', w[0][0])
print('w_1 = ', w[1][0])
# chuẩn bị đường thẳng 
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(4, 24, 2)
y0 = w_0 + w_1*x0

# vẽ đường thẳng và các điểm trên biểu đồ
plt.plot(X, y, 'ro')
plt.plot(x0,y0)
plt.axis([0, 25, -6, 25])
plt.xlabel('Dân số(chục ngàn người)')
plt.ylabel('Lợi nhuận(triệu VND)')
plt.show()
```

Ta có thể thấy các điểm màu đỏ nằm khá gần đường thẳng màu xanh,vậy ô hình Linear Regreesion hoạt động khá tốt với tập dữ liệu.Giờ thì em sẽ sử dụng mô hình này để dự đoán lợi nhuận thu được nếu thành phố đó có 15,5 vạn người sinh sống
